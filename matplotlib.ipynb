```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
xpoints = np.array([0,10])
ypoints = np.array([20,100])

font1 = {'family' : 'Calibri',
          'color':'blue',
           'size' : 20}
font2 = {'family' : 'sans serif',
          'color':'red',
           'size' : 15}

plt.grid()
plt.plot(xpoints , ypoints , 
         'X-.r' ,
         ms = 15 ,
         mec = 'r' ,
         mfc = 'b',
         linewidth = 2)

plt.title('mft' , loc = 'left')
plt.xlabel('mft grade' , fontdict=font1)
plt.ylabel('mft number' , fontdict=font2)

plt.show()
```

```python
lesson = ['python' , 'ML' , 'JAVA' , 'react' , 'node']
number = [200 , 700 , 500 , 200 , 300]
```

```python
plt.plot(lesson , number)
plt.show()
```

```python
plt.scatter(lesson , number)
plt.show()
```

```python
plt.hist(number , bins= 'auto')
```

```python
plt.pie(number , labels= lesson)
```

```python
plt.figure(figsize=(5,5),dpi = 100)
plt.xlabel('lesson')
plt.ylabel('number')
plt.yticks([1000,700,500,200,100])
plt.xticks(rotation=90)

plt.margins(0.7)
plt.text(0,200,'python',fontsize='10')
plt.text(1,700,'ML',fontsize='10')
plt.plot(lesson , number)
plt.show()
```

```python
lesson =['python' , 'java' , 'ML' , 'react']
mft = [900 , 600 , 450 , 100]
sharif = [1000 , 700 , 500 , 200]
```

```python
plt.title('mft and sharif')
plt.xlabel('lesson')
plt.ylabel('number')

plt.plot(lesson , mft , ls='--' , marker='+' , mew = 3)
plt.plot(lesson , sharif , ls='-' , marker='+', lw = 3)

plt.legend(['mft' , 'sharif'] , loc='best')
plt.grid()
plt.annotate('mftpython' , xytext = (0.3 ,500), xy=(0,900) , arrowprops={'facecolor' : 'red' , 'width' : 4}, fontsize =10)
```

```python
plt.subplot(1,2,1)
plt.plot(lesson , mft)
plt.title('MFT')

plt.subplot(1,2,2)
plt.plot(lesson , sharif)
plt.title('sharif')

plt.show()
```
