```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data_year = pd.read_csv('E:\Python video\Data Analaysis (MFT)\Data analysis (MFT)\dataset csv\Age-standardized suicide rates.csv')
data_year
```

```python
data_year.dtypes
```

```python
data_year.isnull().sum()
```

```python
data_year.drop_duplicates(inplace = True)
data_year
```

```python
data_year.replace([' Both sexes', ' Male' , ' Female'] ,  ['Both sexes', 'Male' , 'Female'] , inplace = True)
data_year
```

```python
data_year.drop('Country' , axis = 1 , inplace = True)
data_year
```

```python
data_year_group = data_year.groupby('Sex').mean()
data_year_group
```

```python
plt.bar(data_year_group.columns , data_year_group.loc['Male'] , 0.2 , label = 'Male')
plt.bar(data_year_group.columns , data_year_group.loc['Both sexes'] , 0.3 , label = 'Both sexes' , alpha = 0.7)
plt.bar(data_year_group.columns , data_year_group.loc['Female'] , 0.4 , label = 'Female' , alpha = 0.8)

plt.title('Average suicide rate in 4 different years')
plt.xlabel('year')
plt.ylabel('Suicide rates per 100000 population')
plt.legend()
plt.grid()

plt.show()
```

```python
data_age = pd.read_csv('E:\Python video\Data Analaysis (MFT)\Data analysis (MFT)\dataset csv\Crude suicide rates.csv')
data_age
```

```python
data_age.dtypes
```

```python
data_age.isnull().sum()
```

```python
data_age.drop_duplicates(inplace = True)
data_age.shape
```

```python
data_age.replace([' Both sexes', ' Male' , ' Female'] ,  ['Both sexes', 'Male' , 'Female'] , inplace = True)
data_age
```

```python
data_age_no_country = data_age.drop('Country' , axis = 1)
data_age_no_country
```

```python
df_age = pd.DataFrame(data_age_no_country.groupby('Sex').mean())
df_age
```

```python
plt.plot(df_age.columns , df_age.loc['Male'] , ls = '--', marker = 'o' , mew = 2 ,label = 'Male')
plt.plot(df_age.columns , df_age.loc['Both sexes'] , ls = '--', marker = 'o', mew = 3 , label = 'Both sexes')
plt.plot(df_age.columns , df_age.loc['Female'] , ls = '--', marker = 'o' , mew = 2 , label = 'Female')

plt.title('Suicide')
plt.xlabel('age ranges')
plt.ylabel('Suicide rates per 100000 population')
plt.legend()
plt.grid()

plt.show()
```
