```python
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')

sb.distplot([3,1,2,3,4,5], hist = True)
plt.show()
```

```python
from numpy import random

x = random.normal(loc = 1 , scale = 2 , size = (2,3))
print(x)

sb.distplot(random.normal(size = 1000) , hist = False)
plt.show()
```

```python
y = random.binomial(n = 10 , p = 0.5 , size = 10)
print(y)

sb.distplot(random.binomial(n = 10 , p = 0.5 , size = 1000) , hist = True , kde = True)
plt.show()
```

```python
sb.distplot(random.normal(loc = 50 , scale = 5 ,size = 1000) , hist = False , label = 'normal')
sb.distplot(random.binomial(n = 100 , p = 0.5 , size = 1000) , hist = False , label = 'binomial')

plt.show()
```

```python
x = random.poisson(lam = 2 , size = 10)
print(x)

sb.distplot(random.poisson(lam = 2 , size = 1000) , kde = True)
plt.show()
```

```python
x = random.uniform(size = (2 ,3))
y = random.uniform(20 , 25 , size =  (2,3))
print(x)
print(y)
sb.distplot(random.uniform(size = 1000) , hist = False)

plt.show()
```

```python
from scipy.stats import bernoulli
import seaborn as sb

data_bern = bernoulli.rvs(size = 1000 , p = 0.6)
ax = sb.distplot(data_bern ,
                 kde = True ,
                 color = 'crimson',
                 hist_kws = {
                     'linewidth' : 25 ,
                     'alpha' : 1
                 })

ax.set(xlabel = 'bernoulli' , ylabel = 'frequency')
```
