```python
from scipy.stats import ttest_1samp
import numpy as np
```

```python
ages = np.genfromtxt(r'E:\Python video\Data Analaysis (MFT)\Data analysis (MFT)\dataset csv\ages.csv')
print(ages)

ages_mean = np.mean(ages)
print(ages_mean)

test, pval = ttest_1samp(ages , 30)
print('p-value :' , pval)

if pval < 0.05 :      #alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis ")
else : 
    print("we are accepting null hypothesis")
```

```python
from scipy.stats import ttest_ind

week1 = np.genfromtxt(r'E:\Python video\Data Analaysis (MFT)\Data analysis (MFT)\dataset csv\week1.csv' , delimiter = ",")
week2 = np.genfromtxt(r'E:\Python video\Data Analaysis (MFT)\Data analysis (MFT)\dataset csv\week2.csv' , delimiter = ",")
print(week1)
print("week 2 data : \n")
print(week2)

week1_mean = np.mean(week1)
week2_mean = np.mean(week2)
print("week 1 mean value : " , week1_mean)
print("week 2 mean value : " , week2_mean)

week1_std = np.std(week1)
week2_std = np.std(week2)
print("week 1 std value : " , week1_std)
print("week 2 std value : " , week2_std)

ttest ,pval = ttest_ind(week1 , week2)
print("p-value : " , pval)

if pval < 0.05 :
    print("[***reject***]")
else : 
    print("[***accept***]")
```

```python
import pandas as pd
from scipy import stats

df = pd.read_csv(r'E:\Python video\Data Analaysis (MFT)\Data analysis (MFT)\dataset csv\blood_pressure.csv')
df[['bp_before' , 'bp_after']].describe()
ttest , pval = stats.ttest_rel(df['bp_before'] , df['bp_after'])

print(pval)
if pval < 0.05 :
    print("***REJECT***")
else : 
    print("***ACCEPT***")
```

```python
mydata_house = pd.read_csv(r'E:\Python video\Data Analaysis (MFT)\Data analysis (MFT)\dataset csv\house.csv')
mydata_house
```

```python
mydata_house['Area'] = pd.to_numeric(mydata_house.Area , errors = 'coerce') #change type
mydata_house.replace({False : 0 , True : 1} , inplace = True)  # change type 

mydata_house
```

```python
mydata_house.isnull().sum()
```

```python
mydata_house.drop(index = mydata_house.Address[mydata_house.Address.isnull()].index,inplace= True)   #drop null : Address
mydata_house.shape
```

```python
mydata_house.drop(index = mydata_house.Area[mydata_house.Area.isnull()].index,inplace= True)   #drop null : Area
mydata_house.shape
```

```python
mydata_house.drop_duplicates(inplace=True)
mydata_house.shape
```

```python
mydata_house.reset_index(drop = True , inplace = True)
```

```python
address = set (mydata_house.Address)
Address = {}
for i , j in enumerate (address , start = 1):
    x = {j : i}
    Address.update(x)
number_address = mydata_house.Address.map(lambda s:Address.get(s) if s in Address else s)
number_address
```

```python
address = set(mydata_house.Address)
Address = {}
for i in address :
    j = mydata_house[mydata_house['Address'] == i].Price.sum() / mydata_house[mydata_house["Address"] == i].Area.sum()
    x = {i : j}
    Address.update(x)
avrage_price = mydata_house.Address.map(lambda s:Address.get(s) if s in Address else s)
avrage_price
```

```python
mydata_house.drop(mydata_house[(mydata_house['Area'] > 500.0) &
                (mydata_house['Room'] < 4)].index , axis = 0 , inplace = True)
```
