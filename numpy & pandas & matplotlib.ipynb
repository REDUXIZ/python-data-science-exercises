```python
def ECDF (data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1 , n+1) / n
    return x , y
```

```python
x , y = ECDF (smartphone.inch)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize = (5,5))
plt.scatter(x,y,s = 50)
plt.margins(0.05)

plt.xlabel('inch' , fontsize = 15)
plt.ylabel('ECDF' , fontsize = 15)
```
