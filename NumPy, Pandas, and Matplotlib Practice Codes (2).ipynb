```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
smartphone = pd.read_csv('E:\Python video\Data Analaysis (MFT)\Data analysis (MFT)\dataset csv\smartphone.csv')
smartphone

smartphone.drop(['Name' , 'Company'] , axis = 1 , inplace = True)
smartphone

smartphone_data = pd.get_dummies(smartphone)
smartphone_data
```

```python
smartphone_data.describe()
```

```python
from sklearn.preprocessing import scale , normalize , minmax_scale
```

```python
scale_data = scale(smartphone_data)
```

```python
scale_data
```

```python
df_smartphone = pd.DataFrame(scale_data , index = smartphone_data.index , columns=smartphone_data.columns)
df_smartphone
```

```python
norm_data = normalize(smartphone_data , norm = 'l1' , axis = 0)
norm_data
```

```python
df_smartphone = pd.DataFrame(norm_data , index = smartphone_data.index , columns= smartphone_data.columns)
df_smartphone
```

```python
norm_data = normalize(smartphone_data , norm = 'l2' , axis = 0)
norm_data
```

```python
df_smartphone = pd.DataFrame(norm_data , index = smartphone_data.index , columns= smartphone_data.columns)
df_smartphone
```

```python
minmax_df = minmax_scale(smartphone_data , feature_range = (-1 , 1))
df_smartphone = pd.DataFrame(minmax_df , index = smartphone_data.index , columns= smartphone_data.columns)
df_smartphone
```
